```python
import pandas as pd

# File paths
file_paths = [f"/kaggle/input/bootstrapped-1/BootStrapped-Level-1-filter-Gpt/BootStrapped-Level-1-filter-Gpt/Patient-{i}-results-converted.csv" for i in range(1, 17)]
file_names= [f"Patient-{i}-results-converted.csv" for i in range(1, 17)]
# Function to load, modify, and save each file
def modify_and_save_file(file_path,file_name):
    # Load the file into a DataFrame
    df = pd.read_csv(file_path)

    # Check if 'PubID' column exists
    if 'PubID' not in df.columns:
        raise ValueError(f"'PubID' column not found in {file_path}")

    # Add 'Links' column
    df['Links'] = 'https://pubmed.ncbi.nlm.nih.gov/' + df['PubID'].astype(str)

    # Save the modified DataFrame back to a new CSV file
    new_file_path = file_name.replace("-converted.csv", "-with-links.csv")
    df.to_csv(new_file_path, index=False)
    return new_file_path

# Apply the function to all files
modified_file_paths = [modify_and_save_file(file_path,file_name) for file_path,file_name in zip(file_paths,file_names)]
modified_file_paths
```
